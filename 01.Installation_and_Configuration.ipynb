```
import azureml.core

# Check core SDK version number - based on build number of preview/master.
print("Version Azure ML service : ", azureml.core.VERSION)
print()
import sys
print("Version Python : ", sys.version)
```

```
# Renseigner les informations de votre workspace Azure ML service
subscription_id = "70b8f39e-8863-49f7-b6ba-34a80799550c"
resource_group = "mlserviceresourcegroup"
workspace_name = "MLServiceWorkspace"
workspace_region = "West Europe"
```

```
# On importe les définitions du workspace Azure ML service
# Renseigner le code depuis l'URL indiquée. Vous pouvre ensuite fermer la fenêtre !

from azureml.core import Workspace

ws = Workspace.create(name = workspace_name,
                      subscription_id = subscription_id,
                      resource_group = resource_group, 
                      location = workspace_region,
                      exist_ok=True)

ws.get_details()
```

> Authentification Azure : https://microsoft.com/devicelogin

```
ws = Workspace(workspace_name = workspace_name,
               subscription_id = subscription_id,
               resource_group = resource_group)

# Permet la persistence des infos
ws.write_config()
```
